```python
import torch
import torch.nn as nn
from torch.distributions import Normal
import gym
import numpy as np
import random
from copy import deepcopy
import matplotlib.pyplot as plt
```

```python
import os    
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

```python
class SAC(nn.Module):
    def __init__(self, states_dim, actions_dim, batch_size=64, gamma = .99, alpha = .001, 
                 tau = .01, pi_lr = 1e-3, q_lr = 1e-3):
        super().__init__()
        
        self.gamma = gamma
        self.alpha = alpha
        self.batch_size = batch_size
        self.tau = tau
        
        self.q_net1 = nn.Sequential(nn.Linear(states_dim + actions_dim, 128), 
                                    nn.ReLU(), nn.Linear(128, 128),
                                    nn.ReLU(), nn.Linear(128, 1))
        
        self.q_net2 = nn.Sequential(nn.Linear(states_dim + actions_dim, 128), 
                                    nn.ReLU(), nn.Linear(128, 128),
                                    nn.ReLU(), nn.Linear(128, 1))
        
        self.q_targ1 = deepcopy(self.q_net1)
        self.q_targ2 = deepcopy(self.q_net2)
        
        self.pi_net = nn.Sequential(nn.Linear(states_dim, 128), nn.ReLU(), nn.Linear(128, 128),
                                   nn.ReLU(), nn.Linear(128, 2 * actions_dim), nn.Tanh())
        #модель предсказ мат.ож-е и дисперсию. tanh - makes mean at [-1, 1]
        
        self.memory = []
        self.optimizer_q1 = torch.optim.Adam(self.q_net1.parameters(), q_lr)
        self.optimizer_q2 = torch.optim.Adam(self.q_net2.parameters(), q_lr)
        
        self.optimizer_pi = torch.optim.Adam(self.pi_net.parameters(), pi_lr)
        
    def predict_action(self, state):
        mean, log_std = self.pi_net(state).T
        mean, log_std = mean.unsqueeze(1), log_std.unsqueeze(1) 
        dist = Normal(mean, torch.exp(log_std))
        action = dist.rsample() #rsample  прокидывает градиенты в отличие от sample
        log_prob = dist.log_prob(action)
        return action, log_prob
    
    def get_action(self, state):
        state = torch.FloatTensor(state).unsqueeze(0)
        action, _ = self.predict_action(state)
        return action.squeeze(1).detach().numpy()
    
    def fit(self, state, action, reward, done, n_state):
        self.memory.append([state, action, reward, done, n_state])
        if len(self.memory) > self.batch_size:   
            batch = random.sample(self.memory, self.batch_size)
            states, actions, rewards, dones, n_states = map(torch.FloatTensor, zip(*batch))
            rewards, dones = rewards.unsqueeze(1), dones.unsqueeze(1)
            
            n_actions, n_log_probs = self.predict_action(n_states)
            
            n_states_and_n_actions = torch.concatenate((n_states, n_actions), dim = 1)
            n_q1_values = self.q_targ1(n_states_and_n_actions)
            target = rewards + self.gamma * (1 - dones) * (torch.min(n_q1_values, 
                self.q_targ2(n_states_and_n_actions)) -  self.alpha * n_log_probs)
            
            actions_states = torch.concatenate((states, actions), dim = 1)
            q1_loss = torch.mean((target.detach() - self.q_net1(actions_states)) ** 2)
            q2_loss = torch.mean((target.detach() - self.q_net2(actions_states)) ** 2)
            
            self.update_model(q1_loss, self.optimizer_q1, self.q_net1, self.q_targ1)
            self.update_model(q2_loss, self.optimizer_q2, self.q_net2, self.q_targ2)
            
            actions1, log_probs1 = self.predict_action(states)
            actions_states1 = torch.concatenate((states, actions1), dim = 1)
            pi_loss = torch.mean(torch.min(self.q_net1(actions_states1), 
                self.q_net2(actions_states1))  -  self.alpha * log_probs1)
            self.update_model(pi_loss, self.optimizer_pi)
            
    def update_model(self, loss, optimizer, model = None, target_model = None):
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()    
        
        if model != None and target_model != None:
            for param, target_param in zip(model.parameters(), target_model.parameters()):
                new_target_param = (1 - self.tau) * target_param + self.tau * param
                target_param.data.copy_(new_target_param)
            
```

```python
env = gym.make('Pendulum-v1')
states_dim = env.observation_space.shape[0]
actions_dim = env.action_space.shape[0]

agent = SAC(states_dim, actions_dim)

episode_n = 100
total_reward = []
for _ in range(episode_n):
    total = 0
    state = env.reset()
    
    for t in range(200):
        action = agent.get_action(state)

        n_state, reward, done, _ = env.step(2 * action)

        agent.fit(state, action, reward, done, n_state)

        state = n_state
        total += reward
    total_reward.append(total)
```
